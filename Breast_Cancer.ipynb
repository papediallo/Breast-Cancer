*Breast Cancer Classification with Neural Network* | **Deep Learning**

Predict if a particular patient has benign tumor or malignant 

```python

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

laoding the data 

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

```python
#load the data to data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data , columns = breast_cancer_dataset.feature_names)
```

```python
#print 5 first rows
data_frame.head()
```

adding the target variable which determine if the patient has  benign tumor or malignant


```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
#print last 5 rows 
data_frame.tail()
```

```python
# num of rows and 
data_frame.shape

# getting some info
data_frame.info()
```

```python
# getting some information about missing value
data_frame.isnull().sum()
```

```python
#statitical measures
data_frame.describe()
```

```python
# value of the target variable 
data_frame['label'].value_counts()
```



   1 -> Benign

  0 -> Malignant

Separating the features and target

```python
X = data_frame.drop( columns = 'label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

```python
X_train , X_test, Y_train , Y_test = train_test_split( X, Y, test_size=0.2, random_state= 2)
```

```python
print(X.shape , X_train.shape , X_test.shape )
```

Standarise the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
#normaliser les données en les faisant correspondre à une distribution gaussienne avec une moyenne nulle et une variance unitaire.
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
print(X_train_std)
```

**Neural** **NetWork** 

```python
# importing tensorflow and keras 

#TensorFlow is a powerful and flexible framework developed by Google for building and deploying machine learning models. It provides a wide range of tools and functions for building, training, and evaluating models, as well as tools for visualization and debugging.
#Keras is a high-level neural network library built on top of TensorFlow. It provides a simplified interface for building, training, and evaluating deep learning models. Keras makes it easy to build complex models, such as convolutional neural networks (CNNs) and recurrent neural networks (RNNs), 
import tensorflow as tf 
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of NN

model = keras.Sequential([
                          keras.layers.Flatten(input_shape = (30,)),
                          keras.layers.Dense( 20, activation='relu'),
                          #keras.layers.Dense( 30, activation='sigmoid'),
                          keras.layers.Dense( 2, activation = 'sigmoid') # because we are 2 classes( in the label )

])

```

```python
#from keras import optimizers
# compile the NN
#La fonction de perte mesure l'erreur du modèle en comparant les prédictions du modèle aux valeurs réelles pour les données d'entraînement.

#les métriques sont utilisées pour évaluer la performance du modèle une fois formé

model.compile(optimizer ='adam',
              loss= 'sparse_categorical_crossentropy',
              metrics = ['accuracy']
) 
```

```python
#training the neural NetWork
# validation_split est un paramètre qui spécifie la proportion des données d'entraînement à utiliser pour la validation du modèle. Les données de validation sont utilisées pour évaluer la performance du modèle au fil du temps sans être utilisées pour ajuster les poids du modèle. Le but est d'éviter que le modèle ne surapprenne les données d'entraînement
# "epoch" est le nombre de fois que le modèle va voir l'ensemble des données d'entraînement. Chaque fois que le modèle voit l'ensemble des données d'entraînement, cela est considéré comme une époque.
history = model.fit(X_train_std, Y_train, validation_split = 0.1, epochs = 10)
```

VISUALIZATION **ACCURACY** & **LOSS**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['train_data', 'validation_data'] , loc= 'lower right ')

```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model_accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['train_data', 'validation_data'] , loc= 'upper right ')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate( X_test_std , Y_test)
print(accuracy)
```

Building the predictive system

```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# change this input to np. array

input_data_as_numpy_array = np.array( input_data)

#reshape the numpy array 
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data 
input_data_std = scaler.transform(input_data_reshaped)

#prediction

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [ np.argmax(prediction)]

print(prediction_label)

if(prediction_label[0] == 0 ):
  print("The tumer is Malignant")
else:
  print("The tumer is benign")


```
